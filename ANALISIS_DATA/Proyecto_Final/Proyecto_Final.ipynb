```python
import pandas as pd
```

```python
vehiculos = pd.read_csv("VEHICULOS_INVOLUCRADOS_EN_UN_ACCIDENTE_DE_TRANSITO.csv")
victimas = pd.read_csv("Homicidios_accidente_de_tránsito.csv")
```

```python
victimas.info()
```

```python
# Revisamos el tipo de datos en la columna de edad del vehículo ya que están en diferentes tipos de variables.
vehiculos["EDAD_VEHICULO"].unique()
```

```python
# Separamos el año y el año del accidente para poder calcular de forma acertada la edad del vehiculo al momento del accidente

vehiculos["AÑO_ACCIDENTE"] = vehiculos["FECHA_ACCIDENTE"].str.split("/").str[1].astype(int)
vehiculos["MES_ACCIDENTE"] = vehiculos["FECHA_ACCIDENTE"].str.split("/").str[0].astype(int)
```

```python
# Calculamos la edad del vehículo al momento del accidente, restando el año del accidente al modelo del vehículo

vehiculos["EDAD_VEHICULO"] = vehiculos["AÑO_ACCIDENTE"] - vehiculos["MODELO_VEHICULO"]

# Cambiamos los valores menores a CERO por CERO para evitar errores
vehiculos.loc[vehiculos["EDAD_VEHICULO"] < 0, "EDAD_VEHICULO"] = 0
```

## Análisis exploratorio

```python
# ¿Cuántos accidentes terminaron con muertos y cuántos con heridos?

vehiculos["GRAVEDAD_ACCIDENTE"].value_counts()
```

```python
# ¿Qué porcentaje representa cada uno?

vehiculos["GRAVEDAD_ACCIDENTE"].value_counts(normalize=True)*100
```

```python
# ¿Qué tipos de vehículo aparecen más frecuentemente?

vehiculos["TIPO_VEHICULO"].value_counts().head(10)
```

```python
# ¿Cuáles son los departamentos con más accidentes?

vehiculos["DEPARTAMENTO_ACCIDENTE"].value_counts().head(10)
```

```python
# ¿Cuál es la edad promedio de los vehículos según la gravedad?

vehiculos.groupby("GRAVEDAD_ACCIDENTE")["EDAD_VEHICULO"].mean()
```
